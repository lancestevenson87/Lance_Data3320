```python
while True:
    try:
        apples = int(input('How Many Apples did you eat?'))
        if apples < 1 or apples > 10:
            raise ValueError     # Restarts Question
        break
    except ValueError:
        print("Must be a whole number from 1-10, if your eating more then that see your doctor...")

print("That is", int(apples) * 95, "Calories")

while True:
    try:
        bananas = int(input('How Many Bananas did you eat?'))
        if bananas < 1 or bananas > 10:
            raise ValueError     # Restarts Question
        break
    except ValueError:
        print("Must be a whole number from 1-10, if your eating more then that see your doctor...")

print("That is", int(bananas) * 105, "Calories")

print("Your total calories is", (int(apples)*95)+(int(bananas)*105))
```
